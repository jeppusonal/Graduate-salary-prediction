```python
# <fill_this>

# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define hyperparameter options
fit_intercept_options = [True, False]
n_jobs_options = [-1, 1]  # Number of CPU cores (-1 means using all cores)

# Dictionary to store MSE results for each hyperparameter combination
mse_results = {}

# Manually iterate over each hyperparameter combination
for fit_intercept in fit_intercept_options:
    for n_jobs in n_jobs_options:
        # Initialize the Linear Regression model with the current hyperparameters
        model = LinearRegression(fit_intercept=fit_intercept, n_jobs=n_jobs)

        # Fit the model on the training data
        model.fit(X_train, y_train)

        # Generate predictions on the validation data
        y_pred_val = model.predict(X_val)

        # Calculate the Mean Squared Error on the validation set
        rmse = mean_squared_error(y_val, y_pred_val)

        # Store the MSE score for this combination of hyperparameters
        mse_results[(fit_intercept, n_jobs)] = rmse

# Print MSE for each hyperparameter combination
for hyperparams, mse_score in mse_results.items():
    fit_intercept, n_jobs = hyperparams
    print(f"fit_intercept: {fit_intercept}, n_jobs: {n_jobs} - RMSE: {mse_score:.4f}")

# Identify the hyperparameters with the lowest MSE
best_hyperparams = min(mse_results, key=mse_results.get)
best_fit_intercept, best_n_jobs = best_hyperparams
print(f"\nBest Hyperparameters: fit_intercept = {best_fit_intercept}, n_jobs = {best_n_jobs}")
print(f"Lowest Validation MSE: {mse_results[best_hyperparams]:.4f}")

# Fit the best model with the optimal hyperparameters on the entire training set
best_model = LinearRegression(fit_intercept=best_fit_intercept, n_jobs=best_n_jobs)
best_model.fit(X_train, y_train)

# Predict on the validation set using the best model
final_predictions = best_model.predict(X_val)

# Calculate the final RMSE for the best model on the validation set
final_mse = mean_squared_error(y_val, final_predictions)
print(f"Final Validation RMSE with Best Model: {final_mse:.4f}")
```

```python
# <fill_this>

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# Sample Data Preparation (Replace with your actual data)
# X and y should be the features and target variable from your dataset
X = X_test
y = y_test

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize lists to store results
results = []

# Train and evaluate the model with fit_intercept=True
model_intercept_true = LinearRegression(fit_intercept=True)
model_intercept_true.fit(X_train, y_train)
y_pred_intercept_true = model_intercept_true.predict(X_test)
mse_intercept_true = mean_squared_error(y_test, y_pred_intercept_true)
mae_intercept_true = mean_absolute_error(y_test, y_pred_intercept_true)
results.append({
    'fit_intercept': True,
    'mean_squared_error': mse_intercept_true,
    'mean_absolute_error': mae_intercept_true
})

# Train and evaluate the model with fit_intercept=False
model_intercept_false = LinearRegression(fit_intercept=False)
model_intercept_false.fit(X_train, y_train)
y_pred_intercept_false = model_intercept_false.predict(X_test)
mse_intercept_false = mean_squared_error(y_test, y_pred_intercept_false)
mae_intercept_false = mean_absolute_error(y_test, y_pred_intercept_false)
results.append({
    'fit_intercept': False,
    'mean_squared_error': mse_intercept_false,
    'mean_absolute_error': mae_intercept_false
})

# Print out the results
for result in results:
    print(f"Fit Intercept: {result['fit_intercept']}")
    print(f"Mean Squared Error: {result['mean_squared_error']:.4f}")
    print(f"Mean Absolute Error: {result['mean_absolute_error']:.4f}")
    print()
```

This code compares the performance of Linear Regression models with and without intercepts by evaluating them on test data using both RMSE and MAE to determine which setting provides better accuracy and error metrics.


> Results:

**Findings:**

- **Fit Intercept=True:** Lower RMSE, suggesting better model accuracy with an intercept.
- **Fit Intercept=False:** Higher RMSE, indicating less effective performance without an intercept.
