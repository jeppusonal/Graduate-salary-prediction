### E.1 Import Algorithm

```python
# Import Ridge regression model from scikit-learn for regularization to prevent overfitting and improve model performance
from sklearn.linear_model import Ridge
```

<hr>

### E.2 Set Hyperparameters

```python
# Define a list of alpha values for Ridge regression to test different levels of regularization strength
# These values range from no regularization (0) to strong regularization (1000) to find the optimal balance

alpha_values = [0, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
```

<hr>

### E.3 Fit Model

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize a dictionary to store RMSE results for each alpha value
results = {}

# Train Ridge models for each alpha value
for alpha in alpha_values:
    model_ridge = Ridge(alpha=alpha)

    # Train the model on the training data
    model_ridge.fit(X_train, y_train)

    # Predictions for the training set
    y_pred_train = model_ridge.predict(X_train)

    # Predictions for the validation set
    y_pred_val = model_ridge.predict(X_val)

    # Predictions for the test set
    y_pred_test = model_ridge.predict(X_test)

    # Calculate RMSE for training, validation, and test sets
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

    # Store the results in the dictionary
    results[alpha] = {'train_rmse': rmse_train, 'val_rmse': rmse_val, 'test_rmse': rmse_test}
```

Rationale:
Evaluate Ridge Regression models across various `alpha` values by calculating RMSE for training, validation, and test sets to identify the optimal regularization strength for improved model performance.
