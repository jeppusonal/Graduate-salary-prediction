## E. Train Machine Learning Model

### E.1 Import Algorithm

> Rationale: <fill_this>

```python
# Import Lasso regression model from scikit-learn for feature selection and regularization to improve model performance
from sklearn.linear_model import Lasso
```

<hr>

### E.2 Set Hyperparameters

> Rationale: <fill_this>

```python
# Define a list of alpha values for Lasso regression to test different levels of regularization strength
# These values range from no regularization (0) to strong regularization (100) to find the optimal balance
alpha_values = [0, 0.001, 0.01, 0.1, 1, 1, 10, 100]
```

<hr>

### E.3 Fit Model

```python

from sklearn.metrics import mean_squared_error as mse

# Initialize a dictionary to store results
results = {}

# Train Lasso models for each alpha value
for alpha in alpha_values:
    model_lasso = Lasso(alpha=alpha)

    # Train the model on training data
    model_lasso.fit(X_train, y_train)

    # Predictions on the training set
    y_pred_train = model_lasso.predict(X_train)

    # Predictions on the validation set
    y_pred_val = model_lasso.predict(X_val)

    # Predictions on the test set
    y_pred_test = model_lasso.predict(X_test)

    # Calculate RMSE for training, validation, and test sets
    rmse_train = mse(y_train, y_pred_train, squared=False)
    rmse_val = mse(y_val, y_pred_val, squared=False)
    rmse_test = mse(y_test, y_pred_test, squared=False)

    # Store the results in the dictionary
    results[alpha] = {'train_rmse': rmse_train, 'val_rmse': rmse_val, 'test_rmse': rmse_test}
```

The code here trains several Lasso Regression models for different `alpha` values that define the intensity of regularization. Afterwards, it calculates and stores the RMSE concerning training, validation, and test sets to be able to assess the performance and generalization of the model for different levels of regularization.
